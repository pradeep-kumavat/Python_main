```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
#resize your graph
mp.figure(figsize=(6,3),dpi=300)

#mp.plot(x,y,label='2x',color="red",linewidth="2",linestyle="dashed",marker=".",markersize="22",markeredgecolor="blue")

mp.plot(x,y,"ro--",label='2x')
#line no. number 2----
x2= np.arange(0,4.5,0.5)
mp.plot(x2[:6],x2[:6]**2,"yo--",label="x^2")
mp.plot(x2[4:],x2[4:]**2,"yo-",label="x^2")

mp.title("Our first Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20})
mp.xlabel("X axis")
mp.ylabel("Y axis")
mp.legend()

mp.xticks([0,1,2,3,4])
# mp.yticks([0,2,4,6,8,10])
mp.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
#resize your graph
mp.figure(figsize=(6,3),dpi=100)

#mp.plot(x,y,label='2x',color="red",linewidth="2",linestyle="dashed",marker=".",markersize="22",markeredgecolor="blue")

#short hand notation ---line 1
mp.plot(x,y,"ro--",label='2x')

#line no. 2----
x2= np.arange(0,4.5,0.5)
mp.plot(x2[:6],x2[:6]**2,"yo--",label="x^2")
mp.plot(x2[4:],x2[4:]**2,"yo-",label="x^2")

mp.title("Our first Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20})
mp.xlabel("X axis")
mp.ylabel("Y axis")
mp.legend()

mp.xticks([0,1,2,3,4])
# mp.yticks([0,2,4,6,8,10])

mp.savefig("myfirstgraph.png",dpi=300)
mp.show()
```

# Bar Chart

```python
x= [1,2,3]
y=[1,2,3]
bars = mp.bar(x,y)
#another notation
patterns=['xx' , '*' , '||']
for bar in bars:
    bar.set_hatch(patterns.pop())

#1st notation    
# bars[0].set_hatch('xxxxx')
# bars[1].set_hatch('*')
# bars[2].set_hatch('|||')

mp.figure(figsize=(6,4),dpi=300)

mp.show()
```

# gas price across country in years

```python
gas= pd.read_csv("gas_prices.csv")

mp.figure(figsize=(8,5))
mp.title("Gas prices overtime(in USD)" , fontdict={'fontsize':20, 'fontweight':'bold'})

mp.plot(gas.Year,gas.USA,'b.-',label='United States')
mp.plot(gas.Year,gas.Canada,'g.-',label='Canada')
mp.plot(gas.Year,gas['South Korea'],'r.-',label='South Korea')
# another way to plot many values
# for country in gas:
#     if country!='Year':
#         mp.plot(gas.Year,gas[country],marker='.')

mp.xlabel("Year")
mp.ylabel("US Dollars")
mp.xticks(gas.Year[::2])
mp.yticks()
print(gas.Year[::2])
mp.legend()
mp.savefig('Gaspricesgraph.png',dpi=300)
mp.show()
```

# fifa data

```python
fifa = pd.read_csv("fifa_data.csv")
fifa.head()
```

### histograms

```python
mp.hist(fifa.Overall[:10])
mp.title("fifa data",fontdict={'fontsize':20})

mp.show()
```

# Pie charts

```python
Left = fifa.loc[fifa["Preferred Foot"]== 'Left'].count()[0]
Right = fifa.loc[fifa["Preferred Foot"]== 'Right'].count()[0]
labels=["Left","Right"]
colors=["#abcdef","#aabbcc"]
mp.title("Foot preference of fifa Players")
mp.pie([Left,Right], labels=labels,colors=colors,autopct='%.2f %%')
mp.show()
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight[0]
mp.style.use('ggplot')
```

```python
light= fifa.loc[fifa.Weight < 125].count()[0]
light_medium =fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0]
medium =fifa[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
medium_heavy =fifa[(fifa.Weight>=175) & (fifa.Weight<200)].count()[0]
heavy =fifa.loc[(fifa.Weight>=200)].count()[0]

weights= [light,light_medium,medium, medium_heavy,heavy]
labels= ['Under 125','125-150','150-175','175-200','Over 200']
explode =(.4,.2,0,.0,.4)

mp.title("Weight Distribution of Fifa Players(in lbs)")
mp.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=0.5,explode=explode)
mp.show()
```

```python
mp.style.use("default")
mp.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels=['FC Barcelona','Real Madrid','New England Revolution']

boxes = mp.boxplot([barcelona,madrid,revs],labels=labels,patch_artist=True)

for box in boxes['boxes']:
    #set edge color
    box.set(color='#616087',linewidth=2)
    #set fill color
    box.set(facecolor="#bdafbd")

mp.title("Professional Soccor team Compition")
mp.ylabel("Fifa Overall Rating")

mp.show()
```
